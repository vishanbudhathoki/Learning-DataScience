```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/solar.csv")
```

```python
df.head()
```

```python
plt.boxplot(df['Generation (GWh)'])
```

```python
plt.figure(figsize=(10,5))
plt.barh(df['State'],df['Generation (GWh)'],color=['red','blue','green','pink'])
```

Pie Chart

```python
plt.figure(figsize=(30,10))
plt.pie(df['Generation (GWh)'],labels=df['State'],autopct='%1.1f%%',explode=(0.2,0,0,0,0,0,0,0))
plt.legend(loc='upper right')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(30,10))
plt.pie(df['Generation (GWh)'],labels=df['State'],autopct='%1.1f%%',wedgeprops=dict(width=0.2))
plt.legend(loc='upper right')
plt.axis('equal')
plt.show()
```

Relationship between two variables

1) If both are numerical use scatter plot

```python
plt.scatter(df['Installed Capacity (MW)'], df['Generation (GWh)'])
plt.xlabel('Installed Capacity (MW)')
plt.ylabel('Generation (GWh)')
```
