```python
c=df.dropna(how='all',axis=0,subset=['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g'])
```

```python
c.shape
```

```python
c.isna().sum()
```

```python
sns.displot(data=c,x='culmen_length_mm',kde=True)
```

```python
c['culmen_length_mm']=c['culmen_length_mm'].fillna(c['culmen_length_mm'].median())
```

```python
c.isna().sum()
```

```python
c['sex'].value_counts()
```

```python
c['sex'] = c['sex'].str.strip().str.upper().replace({
    '-':np.nan,
    '.':np.nan,
    'FAMLE':'FEMALE',
    'FEM':'FEMALE',
    'FEMALES':'FEMALE'
})
```

```python
c['sex'].value_counts()
```

```python
c.isna().sum()
```

```python
sex_mode_value = c['sex'].mode()[0]
```

```python
c['sex'] = c['sex'].fillna(value=sex_mode_value)
```

```python
c.isna().sum()
```

```python
c.info()
```

```python
c['species'].value_counts()
```

```python
c['species'] = c['species'].str.strip().str.upper()
```

```python
c['island'].value_counts()
```

```python
c['island'] = c['island'].str.strip().str.upper()
```

Group By and Aggregate Functions

```python
gp_df = c.groupby(by=['species']).agg({"culmen_length_mm":"min", "body_mass_g":"mean"}).reset_index()
```

```python
c[c['species']=='ADELIE']['culmen_length_mm'].min()
```

```python
c[c['species']=='ADELIE']['body_mass_g'].mean()
```

```python
gp_df.head()
```

```python
c.groupby(by=['species']).agg({"culmen_length_mm":"max"}).reset_index()
```

```python
c.groupby(by=['species','sex']).agg({"culmen_length_mm":['min','max','mean','median','count'],"body_mass_g":"mean"}).reset_index()
```

```python
c.head()
```

```python
result_df = c.groupby(by=['species','island','sex']).agg(
    body_mass_minimum = pd.NamedAgg(column="body_mass_g",aggfunc="min"),
    body_mass_mean = pd.NamedAgg(column="body_mass_g",aggfunc="mean") 
).reset_index()
```
